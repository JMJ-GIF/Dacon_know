# Train데이터 불러오기

각 년도별 DataFrame을 리스트에 append합니다.

리스트 인덱스별로

0: 2017년도 데이터   

1: 2018년도 데이터 

2: 2019년도 데이터 

3: 202년도 데이터

```python
import pandas as pd
import numpy as np
from glob import glob
from tqdm import tqdm
import warnings

import matplotlib.pyplot as plt
import seaborn as sns

RANDOM_STATE = 42
np.seed = 42
DATA_PATH = "../data_keywords/v1/"

warnings.filterwarnings(action='ignore')
PATH_2017 = DATA_PATH + "train/keywords_2017.csv"
PATH_2018 = DATA_PATH + "train/keywords_2018.csv"
PATH_2019 = DATA_PATH + "train/keywords_2019.csv"
PATH_2020 = DATA_PATH + "train/keywords_2020.csv"

paths = [PATH_2017, PATH_2018, PATH_2019, PATH_2020]

know_train = [pd.read_csv(path) for path in paths]
```

```python
# read index
i_PATH_2017 = "../data_0115/" + "train/KNOW_2017.csv"
i_PATH_2018 = "../data_0115/" + "train/KNOW_2018.csv"
i_PATH_2019 = "../data_0115/" + "train/KNOW_2019.csv"
i_PATH_2020 = "../data_0115/" + "train/KNOW_2020.csv"

i_paths = [i_PATH_2017, i_PATH_2018, i_PATH_2019, i_PATH_2020]
know_train_idx = [pd.read_csv(path) for path in i_paths]
```

```python
for i in range(4):
    know_train[i].index = know_train_idx[i]['idx']
    know_train_idx[i].index = know_train_idx[i]['idx']
    know_train[i]['knowcode'] = know_train_idx[i]['knowcode']

know_train[3] # 2019년도 샘플
```

# 전처리 

`data_keywords`는 이미 전처리된 데이터  
매우 sparse 한 테이블

```python
years = ['2017', '2018', '2019', '2020']
for i in range(4):
    print(f"{2017+i}: {know_train[i].shape}")
```

# X, y 구분 및 모델 학습

이번 대회에서 맞춰야 할 값은 knowcode입니다.

ID와 knowcode를 제외한 나머지 feature를 X, knowcode를 정답 y로 두어 모델을 학습하였습니다.

베이스라인에서는 의사결정나무와 랜덤포레스트를 선정하였습니다

df.drop(['idx','knowcode'], axis=1)

```python
train_data = {}
for year, df in zip(years, know_train):
    train_data[year] = {'X': df.drop('knowcode', axis=1),
                        'y': df['knowcode']} 
```

```python
train_data['2019']
```

# Extra Trees Classifier

```python
from sklearn.ensemble import ExtraTreesClassifier

et_models = {}

for year in tqdm(years):
    model = ExtraTreesClassifier(n_estimators=100, random_state=RANDOM_STATE, n_jobs=8)
    model.fit(train_data[year]['X'].iloc[:, :], train_data[year]['y'])
    et_models[year] = model
```

## Testset 불러오기

마찬가지로 년도별로 DataFrame으로 불러온 후 리스트에 할당합니다.


```python
TEST_PATH_2017 = DATA_PATH + "test/keywords_2017_test.csv"
TEST_PATH_2018 = DATA_PATH + "test/keywords_2018_test.csv"
TEST_PATH_2019 = DATA_PATH + "test/keywords_2019_test.csv"
TEST_PATH_2020 = DATA_PATH + "test/keywords_2020_test.csv"

TEST_PATHs = [TEST_PATH_2017, TEST_PATH_2018, TEST_PATH_2019, TEST_PATH_2020]

# read index
i_TEST_PATH_2017 = "../data_0115/" + "test/KNOW_2017_test.csv"
i_TEST_PATH_2018 = "../data_0115/" + "test/KNOW_2018_test.csv"
i_TEST_PATH_2019 = "../data_0115/" + "test/KNOW_2019_test.csv"
i_TEST_PATH_2020 = "../data_0115/" + "test/KNOW_2020_test.csv"

i_test_paths = [i_TEST_PATH_2017, i_TEST_PATH_2018, i_TEST_PATH_2019, i_TEST_PATH_2020]
know_test_idx = [pd.read_csv(path) for path in i_test_paths]
```

```python
know_test = [pd.read_csv(path) for path in TEST_PATHs]
for i in range(4):
    know_test[i].index = know_test_idx[i]['idx']

know_test[0].head() # 2017년도 test 샘플
```

```python
know_test[1].head()
```

## 테스트셋 추출 및 학습

ID 값을 제외한 나머지 데이터를 이용하여 모델에 넣어 추론합니다.

```python
test_data = {}
for year, df in zip(years, know_test):
    print(year)
    train_columns = train_data[year]['X'].columns
    # print(train_columns)
    test_data[year] =  {'X': df[train_columns]}
```

```python
for year in years:
    print(year)
    print(f"train: {train_data[year]['X'].shape} test: {test_data[year]['X'].shape}")
```

### ET로 예측

```python
et_predicts = [] 

for year in tqdm(years):
    pred = et_models[year].predict(test_data[year]['X'])
    et_predicts.extend(pred)
```

# 제출
