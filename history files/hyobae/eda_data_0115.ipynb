````python
know_code_list = know_train[0]['knowcode'].unique().tolist()
len(know_code_list)
````

````python
knowcode_to_desc = pd.read_csv('../data_pdf_description/pdf_description_2017.csv', index_col='knowcode').to_dict()['description']
````

````python
know_code_keywords = {}

for know_code in know_code_list:
    df = know_train[0][know_train[0]['knowcode'] == know_code]
    big_txt = df.groupby(['knowcode', 'description'])['text_response'].apply(lambda x: ' '.join(x)).iloc[0]

    word_cnt_dict = {}
    for w in big_txt.split():
        if w in stop_words:
            continue

        if w not in word_cnt_dict.keys():
            word_cnt_dict[w] = 1
        else:
            word_cnt_dict[w] += 1

    l_size = 10
    l = sorted(word_cnt_dict.items(), key=lambda x:x[1], reverse=True)[:l_size]
    
    num_l = [x[1] for x in l]
    num_sum = sum(num_l)
    num_l = [l_size * v / num_sum for v in num_l] # 정규화
    txt_l = [x[0] for x in l]
    
    temp_dict = {}
    for i in range(l_size):
        temp_dict[txt_l[i]] = num_l[i]
    know_code_keywords[know_code] = temp_dict
````

````python
know_code_keywords
````

````python
keywords_df = pd.DataFrame(know_code_keywords).fillna(0)
keywords_df['words'] = keywords_df.index
keywords_df = keywords_df[['words'] + know_code_list].reset_index(drop=True)
keywords_df
````
