# 전처리 

* NB: text_response의 keyword 추출한 값으로 학습 시키고  
* ET, RF: text_response 제외한 column들로 학습

```python
from sklearn.preprocessing import LabelEncoder
years = ['2017', '2018', '2019', '2020']
year_encoder = {}

for year, df in zip(years, know_train):
    print(year)
    encoders = {}
    
    for col in df.columns:
        if col == 'idx':
            print("HEHE")
            continue
        
        try:
            df[col] = df[col].map(int)
        except:
            encoder = LabelEncoder()
            df[col] = df[col].map(str)
            df[col] = encoder.fit_transform(df[col])
            encoders[col] = encoder
            
            
    year_encoder[year] = encoders
```

# X, y 구분 및 모델 학습

이번 대회에서 맞춰야 할 값은 knowcode입니다.

ID와 knowcode를 제외한 나머지 feature를 X, knowcode를 정답 y로 두어 모델을 학습하였습니다.

베이스라인에서는 의사결정나무와 랜덤포레스트를 선정하였습니다

df.drop(['idx','knowcode'], axis=1)

```python
train_data = {}
for year, df in zip(years, know_train):
    train_data[year] = {'X': df.drop('knowcode', axis=1),
                        'y': df['knowcode']} 
```

```python
train_data['2017']
```

# Voting Classifier

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import VotingClassifier, ExtraTreesClassifier, RandomForestClassifier

gnb_models = {}
et_models = {}
rf_models = {}

for year in tqdm(years):
    gnb_model = GaussianNB()
    gnb_model.fit(train_data[year]['X'].iloc[:, :], train_data[year]['y'])
    gnb_models[year] = gnb_model

    et_model = ExtraTreesClassifier(n_estimators=100, random_state=RANDOM_STATE, n_jobs=8)
    et_model.fit(train_data[year]['X'].iloc[:, :], train_data[year]['y'])
    et_models[year] = et_model

    rf_model = ExtraTreesClassifier(n_estimators=100, random_state=RANDOM_STATE, n_jobs=8)
    rf_model.fit(train_data[year]['X'].iloc[:, :], train_data[year]['y'])
    rf_models[year] = rf_model
```
