# Null 값에 대하여
 - 공백을 0으로 바꿔서 채워줍시다
 - 존재하는 nan값 역시도 0으로 바꿔서 채워줍시다

```python
# 연도별로 텍스트로 명시된 문항

years = ['2017','2018','2019','2020']

text_2017_question = ['bq4_1a','bq4_1b','bq4_1c','bq5_2','bq19_1','bq30','bq31','bq32','bq33','bq34','bq38_1']
text_2018_question = ['bq4_1a','bq4_1b','bq4_1c','bq5_2','bq28_1','bq29','bq30','bq31','bq32','bq33','bq37_1']
text_2019_question = ['bq4_1a','bq4_1b','bq4_1c','bq5_2','bq18_10','bq20_1','bq22','bq23','bq24','bq27_1']
text_2020_question = ['bq4_1a','bq4_1b','bq4_1c','bq5_2','bq18_10','bq20_1','bq26_1']
whole_list = [text_2017_question, text_2018_question, text_2019_question, text_2020_question]

text_dict = {}
for year, lst in zip(years, whole_list):
    text_dict[year] = lst
```

```python
# train set에 대하여 공백이 아닌 null 값이 존재함 : 확인해보자

null_train_dict = {}
for year, df in zip(years, know_train):
    null_sum = df.isnull().sum()
    null_train_dict[year] = list(null_sum[null_sum!=0].index)    
    print(year,'의 null 합계 :',null_sum.sum())
    print()

# null 값이 나타난 문항들
print(null_train_dict)
```

```python
# test set에 대하여 공백이 아닌 null 값이 존재함 : 확인해보자

null_test_dict = {}
for year, df in zip(years, know_test):
    null_sum = df.isnull().sum()
    null_test_dict[year] = list(null_sum[null_sum!=0].index)    
    print(year,'의 null 합계 :',null_sum.sum())
    print()

# null 값이 나타난 문항들    
print(null_test_dict)
```

```python
# null이 나타난 문항 중에서 text를 요구하는 문항이 많았나?_train
for year in years:
    print(year)
    print('null 값이 나타난 문항 수 :',len(null_train_dict[year]))
    print(set(null_train_dict[year]).intersection(set(text_dict[year])))
    print()
```

```python
# null이 나타난 문항 중에서 text를 요구하는 문항이 많았나?_test
for year in years:
    print(year)
    print('null 값이 나타난 문항 수 :',len(null_test_dict[year]))
    print(set(null_test_dict[year]).intersection(set(text_dict[year])))
    print()
```

```python
# nan 값을 0으로 대체합시다
for df in know_train:
    df.fillna('0',inplace=True)

for df in know_test:
    df.fillna('0',inplace=True)
```

Nan 값의 개수가 적다. 또한 연도별로 Nan 값의 패턴은 보이지 않는 듯. text라고 해서 null 값이 무조건 나타나는 것도 아닌듯 하다

다른 값으로 대체하거나 drop하여도 성능 상에서 별 문제가 되지 않을까 싶다.

```python
# 공백이 주로 어떤 문항에서 많이 등장하는가? train

blank_train_dict = {}
for year, df in zip(years,know_train):
    blank_dist = pd.DataFrame(index=['blank_sum'])
    for col in df.columns:
        blank_dist[col] = df[df[col]==' '].shape[0] 
    blank_train_dict[year] = blank_dist
    
# 공백이 주로 어떤 문항에서 많이 등장하는가? test

blank_test_dict = {}
for year, df in zip(years,know_test):
    blank_dist = pd.DataFrame(index=['blank_sum'])
    for col in df.columns:
        blank_dist[col] = df[df[col]==' '].shape[0] 
    blank_test_dict[year] = blank_dist
```

설문지와 공백의 개수를 비교해보면서 보았는데, 공백이 드러난 부분은 문항에서 필요에 따라 건너뛰라고 지시한 부분이었다.
혹은 text로 작성하는 문항이었다.
blank_sum이 0으로 나타나는 문항들이 꽤 많은데, 이 부분은 참가자가 반드시 마킹을 해야하는 부분이다.
즉, 참여자들이 설문지를 굉장히 모범적으로 작성했으며 설문지 수합 데이터에는 큰 이상이 없다고 말할 수 있겠다.

**Baseline에서 나타났던 공백을 0으로 채우는 부분은 좋은 아이디어인듯! 설문지에 0이라는 응답보기는 없음!**

```python
# train set에 대하여 공백은 모두 str 0으로 전환
for df in know_train:
    for col in df.columns:
        df[col].replace(' ', '0', inplace=True)
```

```python
# test set에 대하여 공백은 모두 str 0으로 전환
for df in know_test:
    for col in df.columns:
        df[col].replace(' ', '0', inplace=True)
```

```python
# int type으로 바꾸는 것을 시도해보고 안된다면 str로 취급하여 정리해봅시다 
print('train')
for year, df in zip(years, know_train):
    print(year)
    
    for col in df.columns:
        try:
            df[col] = df[col].map(int) 
        except:
            df[col] = df[col].map(str)
print()
print('test')
for year, df in zip(years, know_test):
    print(year)
    
    for col in df.columns:
        try:
            df[col] = df[col].map(int) 
        except:
            df[col] = df[col].map(str)
```

```python
def check_datatype_integrity(years, know_data, verbose=True):
    
    trouble_candidate_dict = {}
    
    # 데이터 타입 변환이 잘 이루어졌는지 체크하고 문제가 있는 문항을 딕셔너리에 담습니다
    for year,df in zip(years,know_data):
        compare_a = set(df.dtypes[df.dtypes=='object'].index)
        compare_b = set(text_dict[year])
        if verbose == True:
            print(year)
        if (compare_a - compare_b) == set():
            if verbose == True:
                print('데이터 타입 변환이 잘 이루어 졌습니다')
            trouble_candidate_dict[year] = (compare_a - compare_b)
        else:
            if verbose == True:
                print('문제가 생겼습니다 아래 문항들의 데이터 타입을 확인해보세요')
                print(compare_a - compare_b)
            trouble_candidate_dict[year] = (compare_a - compare_b)
    if verbose == True:
        print('-'*20)    
    
    # 문제가 있는 문항을 출력해서 하나씩 살펴봅시다
    trouble_dict = {}
    
    for year in years:
        trouble_dict[year] = []
        
    for year,df in zip(years,know_data):
        if verbose == True:
            print(year,'년 오류문항 Overview')
            for col in trouble_candidate_dict[year]:
                print(col,':',df[col].unique())
            print()
        for col in trouble_candidate_dict[year]:
            try:
                df[col] = df[col].map(float) # float으로 바꿀 수 있다면 바꿔주고
            except:
                trouble_dict[year].append(col) # 오류가 있다면 딕셔너리에 담아줍시다
    
    return trouble_dict
                
        
```
